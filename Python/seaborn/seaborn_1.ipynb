# seaborn use to data visualization

```python
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### seaborn have some in build data set 

```python
df = sb.load_dataset("tips")
df 
```

```python
sb.relplot(data=df,x = "total_bill", y = "tip",col = "time",hue = "smoker",style = "smoker",size = "size")
# hue="smoker"

# Controls color grouping.

# Seaborn assigns different colors based on unique values in the smoker column.

# style="smoker"

# Controls marker shape.

# Seaborn uses different shapes based on the smoker column.

# size="size"

# Controls marker size.

# Each pointâ€™s size changes based on the numeric values in the size column.
```

```python
iris_df = sb.load_dataset("iris")
iris_df
```

### scatter plot


```python
sb.scatterplot(x = "sepal_length",y = "petal_length",hue="species",data=iris_df)
```
