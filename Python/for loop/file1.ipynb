`isnumeric()` checks if **all characters in a string are numeric**.

```python
a = "123"
print(a.isnumeric())   # True

b = "12a"
print(b.isnumeric())   # False
```

It works **only with strings**, not integers.

````python
list = [1,2,3,6.0,"ss","gt", True, False]
list_int = []
list_str = []
list_bool = []
for i in list:
    if (isinstance(i , int) and isinstance(i , bool)==False):
        list_int.append(i)
    if isinstance(i , str):
        list_str.append(i)
    if isinstance(i , bool):
        list_bool.append(i)
else:
    print(list_int)
    print(list_str)
    print(list_bool)        
                
````

````python
dict = {"name" : "krittik" , "age" : 22,"occupation": ["data eng","student"]}
for i in dict.items():
    print(i)
print("\n")    
for i in dict.keys():
    print(i)
print("\n")      
for i in dict.values():
    print(i)  
print("\n")            
````
