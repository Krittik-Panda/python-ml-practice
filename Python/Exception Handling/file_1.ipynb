### handle valid inplut fromat 

```python
try :
    a = int(input("enter a divisor :"))
    b = 9
    print(b/a)
except :
    print("wrong input !!")

# here we take the vale of b is 0 and any atring it will not be a valid input and will through a error to handle it we can have to use except
```

### if we dont use try except block ...after detect error the program will crash...but if we use try except then the error message from except block will run and after that the rest of the program will also executed ... not crash the whole program 

```python
# give user chance to re enter the number again in except block we call the function again
def take_input():
    try:
        a = int(input("enter the number :"))
        b = 76
        return b/a
    except:
        print("re enter")
        return take_input()
    
print(take_input())    
```

# finally

##### finally will always run

```python
try :
    a = int(input("enter a divisor :"))
    b = 9
    print(b/a)
except :
    print("wrong input !!")
finally:
    print("this will executed always ")    
```

##### print the exact error message ... to do that use Exception class

```python
try :
    a = int(input("enter a divisor :"))
    b = 9
    print(b/a)
except Exception as e:
    print("wrong input !! :",e)
```

#### use sub classes of Exception super class, before using Super class  so that debug will be helpful if any sub classes can't handle the exception then at last Exception class will handle it  

```python
try:
    a = int(input("enter an integer"))
    b = 7 / a
    print(b)

except ValueError as e:
    print("this is wrong input : ", e)

except ZeroDivisionError as e:
    print("this is wrong input : ", e)

except Exception as e:
    print("this is wrong input : ", e)

finally:
    print("this will be executed always ")
```
