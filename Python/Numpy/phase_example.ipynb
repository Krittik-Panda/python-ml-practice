```python
#minimum sales or 1st row 

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],   # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],   # Beijing Bites
    [3, 200000, 230000, 260000, 300000],   # Pizza Hub
    [4, 180000, 210000, 240000, 270000],   # Burger Point
    [5, 160000, 185000, 20500, 230000]    # Chai Point
])

min_sale = np.min(sales_data[0,1:5])
print(min_sale)
```

```python
#min sales per resturent
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],   # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],   # Beijing Bites
    [3, 200000, 230000, 260000, 300000],   # Pizza Hub
    [4, 180000, 210000, 240000, 270000],   # Burger Point
    [5, 160000, 185000, 20500, 230000]    # Chai Point
])

min_sale = np.min(sales_data[:,1:],axis=1)
print(min_sale)
#max sales per year 
max_sale = np.max(sales_data[:,1:],axis=0)
print(max_sale)
#avg sales per resturent
avg_sales = np.average(sales_data[:,1:],axis=1)
print(avg_sales)
#cumulative sums
print("\n\n\n")
cum_sum = np.cumsum(sales_data[:,1:],axis=1)
print(cum_sum)
```

# DATA PLOTING

```python
plt.figure(figsize=(8, 6))
plt.plot(np.mean(cum_sum, axis=0))
plt.title("Average Cumulative Sales Across All Restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

# MULTIPLICATION

```python
vector1 = np.array([2,5,6,7,2,3])
vector2 = np.array([1,9,6,3,2,5])

# addition 
print(f"Addition is :{vector1+vector2}")
print("\n\n")

#multiplication
print(f"Multiplication is :{vector1*vector2}")
print("\n\n")

#dot product
print(f"Dot Product is :{np.dot(vector1,vector2)}")
print("\n\n")

#multiply every elment with 10 in the array 
m = vector1 * 10
print(m)
```
