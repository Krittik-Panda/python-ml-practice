## NUMPY ARRAY OPERATION

```python
import numpy as np
```

```python
arr =np.arange(50)
#slicing
print("slicing :",arr[3:7])
# slicing with step 
print("slicing :",arr[-3:-34:-3])

#negetive indexing 
print("slicing :",arr[-4]) # print 4th element from last
```

```python
arr_2d = np.array([[1,2,3],
                   [3,2,1],
                   [7,5,9]
                    ])
#print specific element 
print("The specific element is in 2nd row 1st column is : ",arr_2d[1,0])
```

```python
arr_2d = np.array([[1,2,3],
                   [3,2,1],
                   [7,5,9],
                   [3,8,5]
                    ])
#print entire row
print("The entire row  : ",arr_2d[2]) #2nd row 
print("The entire column  : ",arr_2d[:,1]) #2nd column 
```

# SORTING

```python
unsorted_array = np.array([4,3,5,6,1,7,8,9,3])
sorted_array = np.sort(unsorted_array)
print(f"the sorted array is : {sorted_array}")
```

```python
unsorted_array_2d = np.array([[1,2,3],
                   [3,2,1],
                   [7,5,9],
                   [3,8,5]
                    ])
# sort 2d array by column use axis value 0
unsorted_array_2d = np.sort(unsorted_array_2d,axis=0)
print(f"the sorted array is : {unsorted_array_2d}")
```

```python
unsorted_array_2d = np.array([[1,2,3],
                   [3,2,1],
                   [7,5,9],
                   [3,8,5]
                    ])
# sort 2d array by row use axis value 1
unsorted_array_2d = np.sort(unsorted_array_2d,axis=1)
print(f"the sorted array is : {unsorted_array_2d}")
```

## FILTER ARRAY

```python
numbers = np.array([2,4,3,4,5,6,4,3,6])
print(f"the element in 4th index is :{numbers[3]}")
#print the even numbers in the array
even_numbers = numbers[numbers%2==0]
print("even elements are :",even_numbers)
sort_even_numbers = even_numbers.sort()
print(f"even elements are sorted :{np.sort(even_numbers)}")
```

#### MASK

```python
numbers = np.array([32,4,37,4,5,6,4,3,6,34,23,45,12,78,56,34,23,58,51,27,93])
mask = numbers >= 10
print(f"numbers that are greater than equal to 10 :{numbers[mask]}")
```

# fancy indexing vs np.where

```python
numbers = np.array([32,4,37,4,5,6,4,3,6,34,23,45,12,78,56,34,23,58,51,27,93])
indices = [0,5,6]
print(numbers[indices])
```

```python
numbers_array = np.array([32,4,37,4,5,6,4,3,6,34,23,45,12,78,56,34,23,58,51,27,93])
mask = np.where(numbers_array > 20) # stores the indices where the condition is true
print("the numbers are : ",numbers_array[mask])
```

##### condition

```python
numbers_array = np.array([32,4,37,4,5,6,4,3,6,34,23,45,12,78,56,34,23,58,51,27,93])
condition_array = np.where(numbers_array >=20 , numbers_array * 2 ,numbers_array)
print("the numbers are : ",condition_array)
```

# ADDING AND REMOVING

```python

original = np.array([[1, 2], [3, 4]])
new_row = np.array([[5, 6]])

with_new_row = np.vstack((original, new_row))
print(original)
print("the new array with new row is :\n",with_new_row)

new_col = np.array([[7], [8]])
with_new_col = np.hstack((original, new_col))
print("With new column :\n", with_new_col)
```
