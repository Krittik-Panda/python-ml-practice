Array Reshape

```python
arr = np.arange(34)
print(f"the array is : \n{arr}")

reshaped = arr.reshape((2,17))
print(f"the  reshaped array is : \n{reshaped}")

flattened = reshaped.flatten()
print(f"the array is : \n{flattened}")

#ravel() return view of reshaped array instead of copy or ravel dont change the array just return the view 
raveled = reshaped.ravel()
print(f"the array is : \n{raveled}")

transposed = reshaped.T
print(f"the array is : \n{transposed}")
 
```
