# __**Types of Argument**__
### __default argument__
### __positional argument__
### __keyword argument__

__---------------------------------------------------------------------------------------------------------------------------------------------__

### 1. Default Argument

A default argument has a predefined value. If the caller does not pass that argument, the default is used.

```python
def greet(name="Unknown"):
    """this function return a greeting message for any name"""
    print(f"Hi {name}")
```

```python
greet("preet")
greet()
```

### 2. Positional Argument

Arguments are assigned based on their position in the function call.

```python
def introduce(name, age):
    return f"{name} is {age} years old"

print(introduce("Sam", 25))   # correct
print(introduce(25, "Sam")) # logically wrong, no error
```

### 3. Keyword Argument

Arguments are passed using parameter names, so order does not matter.

Key idea: improves clarity and avoids ordering mistakes.

```python
def introduce(name, age):
    return f"{name} is {age} years old"

print(introduce(age=25, name="Sam"))  # order doesn't matter
```

# **args* in python

### *args allows a function to accept a variable number of __non keyword__ arguments. These arguments are received as a tuple.

```python
def multiply(*args):
    product = 1
    for i in args:
        product = product * i
    return product

### not necessery to use args its a name of the tuple you can use anything like *alex , *bimal etc        
```

```python
print(multiply(1,2,3,4))
print(multiply(1,2,3,4,5,6,7,8,9))
```

## *kwargs in Python

### *kwargs allows a function to accept a variable number of __keyword__ arguments. These are stored as a dictionary.

```python
def display_info(**kwargs):
    for(key,value) in kwargs.items():
        print(key,'->', value)
### dicstionary
```

```python
display_info(india="delhi",srilanka="colombo",japan="tokyo")
```
