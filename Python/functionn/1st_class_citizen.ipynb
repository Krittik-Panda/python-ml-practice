# Functions in python are first class citizen

### __In Python, the term first-class citizen (or first-class object) most commonly refers to functions. This status means that functions are treated like any other data type, such as integers or strings, and can be manipulated in the same ways.__

```python
def sq(num):
    return num**2
    
```

```python
sq(9)
```

```python
print(type(sq))
print(id(sq))
```

#### reassign

```python
x = sq
x(9)   # now x is the sq() function
```

```python
print(type(x))
```

#### delete a function

```python
del sq
```

#### Store

```python
L =[1,2,3,sq]
L
```

```python
L[-1](4)
```
