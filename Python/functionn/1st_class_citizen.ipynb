### In Python, a higher-order function is a function that either takes one or more functions as arguments or returns a function as its result

```python

def suum():  ### HOF
    def p(r,s):
        return r+s
    return p

suum()(2,3) 
```

```python
def greet():
    return "Hi"

def outer():
    return greet

print(outer())
```

**Case 1:**
`print(outer())` prints something like `<function greet at 0x...>` because `outer()` does not execute `greet`, it only returns a reference to the function object `greet`. When Python prints a function object, it shows its type and memory address instead of running it. Your mistake would be assuming that returning a function automatically calls it—it does not; it just hands you the function itself.



```python
def greet():
    return "Hi"

def outer():
    return greet

outer()
```



**Case 2:**
Calling `outer()` alone produces `<function __main__.greet()>` (or similar) because the returned function is still not being executed—you're just invoking `outer`, which returns `greet`, and the interpreter displays that returned object. The key point is that `outer()` and `greet()` are two separate calls; stopping at `outer()` means you never actually run the inner logic of `greet`.




```python
def greet():
    return "Hi"

def outer():
    return greet

v = outer()
print(v())
```



**Case 3:**
`v = outer(); print(v())` works and prints `Hi` because now you are explicitly calling the returned function. First, `outer()` returns the function `greet` and assigns it to `v`, then `v()` executes it. This is the correct mental model: returning a function gives you a callable object, and you must call it separately to get its result.

#### create a HOF that takes a list and a function (that calculate the square of each element in the list) as argument  

```python
def sq(a):
    return a**2

def transform(func,L):
    output = []
    for i in L:
        output.append(func(i))
    return output

Li = [1,2,3,4,5,6]
transform(func=sq,L=Li)    ### keyword argument
```

##### fulfill the same requirements with lambda function

```python
def transform(func,L):
    output = []
    for i in L:
        output.append(func(i))
    return output

Li = [1,2,3,4,5,6]
transform(func=lambda x:x**2,L=Li)  
```

## Map , Filter , Reduce  -> functions

### Map ->  Applies a function to each item in an iterable(List) and returns a new iterator(List).

```python
listt = list(map(lambda x: x**2,[1,2,3,4]))
print(listt)
```

```python
users = [
    {'name': 'Rahul', 'age': 45, 'gender': 'male'},
    {'name': 'Nitish', 'age': 33, 'gender': 'male'},
    {'name': 'Ankita', 'age': 50, 'gender': 'female'}
]
list(map(lambda u : u['name'],users))
```

### Filter -> Keeps elements where function returns True for a given condition

```python
L = [1,2,3,'y',9,'oo']
list(filter(lambda x: isinstance(x,int),L)) ### filter only numeric value 
```

```python
L = [1,2,3,'y',9,'oo']
list(filter(lambda x: isinstance(x,str),L)) ### filter only character value 
```

### Reduce -> Reduces a list into a single result by repeatedly applying a function

```python
from functools import reduce

nums = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, nums)
print(result)  # 10
```

```python
### find minimum
from functools import reduce

nums = [8, 2, 3, 4]

result = reduce(lambda x, y: x if x<y else y, nums)
print(result)  # 10
```
