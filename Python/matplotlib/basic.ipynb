* `pyplot` is a high-level plotting interface in Matplotlib.
* It provides a MATLAB-style, state-based plotting system.
* It automatically manages figures and axes internally.
* It keeps track of the current active plot using global state.
* It allows quick and simple creation of visualizations.
* It is commonly used in notebooks and small scripts.
* It reduces the need to manually create figure and axis objects.
* It wraps Matplotlib’s underlying object-oriented architecture.
* It includes functions for line plots, scatter plots, histograms, bar charts, labels, titles, and display.
* It is convenient but less explicit in structure.
* It can become harder to manage in complex or large-scale plotting scenarios.
* For more control and scalability, Matplotlib’s object-oriented API is preferred.
* Internally, Matplotlib is structured around Figure, Axes, and Artists objects.

````python
import math
x1 = np.arange(-10, 10,1)
y1 = []
for i in x1:
    y1.append(1/(1+math.exp(-i)))
````

````python
plt.plot(x1,y1)
plt.xlabel("z values")
plt.ylabel("prediction score")
plt.title("sigmoid curve")
plt.show()
````

### plot the values only

````python
plt.plot(x1,y1,'r.')
# plot data in red(r) colour in '.'
plt.show()
````

### multiple plot

````python
plt.plot(x1,y1)
plt.plot(x1,np.sin(y1),'g.')
plt.plot(x1,np.tan(x1))
plt.show()
````

### barplot

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# [0,0,1,1] are coordinates of the rectengle of the bar chart (0,0) and (1,1)...bottom left coordiante is (0,0) top right co ordinate (1,1)
ax.bar(x1,y1)
plt.xlabel("x1")
plt.ylabel("y1")
plt.title("sigmoid bar")
plt.show()
````



### Version 2

```python
plt.bar(x1, y1)
```

You’re saying:

> “Matplotlib, just put this bar chart somewhere.”

Matplotlib decides where.

---

### Version 1

```python
fig, ax = plt.subplots()
ax.bar(x1, y1)
```

You’re saying:

> “Create a specific drawing area. Now draw the bars **inside this exact area**.”

You decide where.

---

## Why your graphs look identical

Because you only have:

* One figure
* One plot
* No layout control

So whether Matplotlib decides…
or you decide…

It ends up the same.

---

## When it actually matters

If you do this:

```python
fig, ax = plt.subplots(1, 2)
ax[0].bar(x1, y1)
ax[1].bar(x2, y2)
```

Now you **must** use `ax.bar()`.

`plt.bar()` would get confused about which subplot to use.

---

## Ultra-simple summary

```
plt.bar() → “Draw it somewhere.”
ax.bar()  → “Draw it here.”
```



### pi chart

````python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
#autopct controls how the percentage text looks on the pie chart.%1.1f%% means one digit after decimal ,%1.2f%% means two digit after decimal
plt.show()
````

### scatter plot

````python

x = np.linspace(0,10,100)
# get evenly spaced 100 values btw 0 to 10
y = np.sin(x)
z = np.tan(y)
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
ax.scatter(y,z,color="g")
#autopct controls how the percentage text looks on the pie chart.%1.1f%% means one digit after decimal ,%1.2f%% means two digit after decimal
plt.show()

````

### 3d plot 

````python

x = np.linspace(0,10,100)
# get evenly spaced 100 values btw 0 to 10
y = np.sin(x)
z = np.tan(y)
fig1 = plt.figure()
ax = plt.axes(projection="3d")
ax.scatter(x,y,z,color="b")
#autopct controls how the percentage text looks on the pie chart.%1.1f%% means one digit after decimal ,%1.2f%% means two digit after decimal
plt.show()

````
