* `pyplot` is a high-level plotting interface in Matplotlib.
* It provides a MATLAB-style, state-based plotting system.
* It automatically manages figures and axes internally.
* It keeps track of the current active plot using global state.
* It allows quick and simple creation of visualizations.
* It is commonly used in notebooks and small scripts.
* It reduces the need to manually create figure and axis objects.
* It wraps Matplotlib’s underlying object-oriented architecture.
* It includes functions for line plots, scatter plots, histograms, bar charts, labels, titles, and display.
* It is convenient but less explicit in structure.
* It can become harder to manage in complex or large-scale plotting scenarios.
* For more control and scalability, Matplotlib’s object-oriented API is preferred.
* Internally, Matplotlib is structured around Figure, Axes, and Artists objects.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(3,23,3)
y = np.arange(7)
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
import math
x1 = np.arange(-10, 10,1)
y1 = []
for i in x1:
    y1.append(1/(1+math.exp(-i)))
```

```python
plt.plot(x1,y1)
plt.xlabel("z values")
plt.ylabel("prediction score")
plt.title("sigmoid curve")
plt.show()
```

### plot the values only

```python
plt.plot(x1,y1,'r.')
# plot data in red(r) colour in '.'
plt.show()
```

### multiple plot

```python
plt.plot(x1,y1)
plt.plot(x1,np.sin(y1),'g.')
plt.plot(x1,np.tan(x1))
plt.show()
```
