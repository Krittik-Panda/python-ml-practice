### ATM Class implementastion 

to declear any variable inside the class ..it always done inside a special methode __init__
init is a constructor  

```python
class ATM:
    def __init__(self):
        self.pin = ""
        self.balance = 0
        self.menu()
    def menu(self):        
        choice = int(input("""
            Enter the option no : 
            1) Set pin
            2) Check balance
            3) Money deposite
            4) Money withdraw
            5) Exit
            """) ) 
        if(choice==1):
            self.set_pin()
            self.menu()
        elif(choice==2):
            self.check_balance()
            self.menu()
        elif(choice==3):
            self.deposite()
            self.menu()
        elif(choice==4):
            self.withdraw()
            self.menu()     
        else:
            print("Thank You !")       
                

    def check_pin(self):
        temp = input("Enter the pin")
        if temp!=self.pin:
            return 0
        else:
            return 1    


    def set_pin(self):
        self.pin = input("Enter your pin ")
        print("Pin is successfully set ")


    def check_balance(self):
        if(self.check_pin()==0):
            print("Wrong pin ") 
        else:
            print(f"the balance of your current account is {self.balance}")  


    def deposite(self):
        try:
            if(self.check_pin()==0):
                print("Wrong pin ")
            else:
                temp = int(input("Enter the deposite amount :"))
                self.balance = self.balance + temp
                print(f"your balance is {self.balance}")
        except:
            print("Enter the digit properly !!!")            


    def withdraw(self):
        try:
            if(self.check_pin()==0):
                print("Wrong pin ")
            else:
                temp = int(input("Enter the withdraw amount : "))
                if(temp>self.balance):
                    print("Insaficient Balance .")
                else:
                    self.balance = self.balance - temp
                    print("Withdrawn is successful ")
                    print(f"The current balance is {self.balance}")    
        except:
            print("Enter digit properly !!!")            
```
