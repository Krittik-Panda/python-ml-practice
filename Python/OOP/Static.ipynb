### aggrigation is a "has a" relation like a customer class have a "has a" relation with address class ... coz every custoner have a address 

```python
class Customer:
    def __init__(self,name,address):
        self.name = name
        self.address = address
class Address:
    def __init__(self,city,pin,state):
        self.city = city
        self.pin = pin
        self.state = state
add = Address("jaipur",123456,"Rajastan")
cust = Customer("muku",add)

print(cust.name)
print(cust.address.city)
```

```python
class Customer:
    def __init__(self,name,address):
        self.name = name
        self.address = address

    def change_profile(self,new_name,new_city,new_pin,new_state):
        self.name = new_name
        self.address.change_address(new_city,new_pin,new_state)    
class Address:
    def __init__(self,city,pin,state):
        self.city = city
        self.pin = pin
        self.state = state

    def change_address(self,new_city,new_pin,new_state):
        self.city = new_city
        self.pin = new_pin
        self.state = new_state

add = Address("jaipur",123456,"Rajastan")
cust = Customer("muku",add)

print(cust.name)
print(cust.address.city) ### ***

cust.change_profile("muku","haldia",721635,"WB")
print(cust.name)
print(cust.address.city)
```

### inheritance is "is a" relationship ..like all car class can inharite all features from vehical class ...coz car "is a" vehical

### a child class inherite all variables ...methodes ...and constructors from the parent class except the private members ...__private members are not inherited__

```python
#### inheritance

class User:
    def login(self):
        print("log in successful")
    def password(self):
        print("password created")


class Student(User):
    ### student is a child class of user
    def register(self):
        print("registered")   


student1 = Student()
student1.login()
student1.password()
student1.register()                 
```

### if child class dont have any constructor it will call the parent class constructor

```python
class Phone:
    def __init__(self,name,price):
        print("inside phone constructor")
        self.name= name
        self.price = price

class Apple(Phone):
    pass
apple = Apple(name = "apple",price = 123410)
print(apple.price)                                             
```
