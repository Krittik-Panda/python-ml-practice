### Reference Variable
#### it is a variable that stores the reference (memory address) of an object, not the object itself.

#### here sbi is the referance variable for the object of ATM



**Name mangling** is a mechanism in Python where any attribute or method with a double underscore prefix (`__name`) is internally renamed to avoid accidental conflicts, especially in inheritance.

---

### What Python actually does

Inside a class:

```python
class ATM:
    def __init__(self):
        self.__pin = 1234
```

Python rewrites it to:

```
_ATM__pin
```

So:

* `__pin` → `_ATM__pin`

---

### Minimal working example

```python
class ATM:
    def __init__(self):
        self.__pin = 1234

atm = ATM()

print(atm.__pin)        # ❌ AttributeError
print(atm._ATM__pin)   # ✅ 1234
```

---

### Why name mangling exists (real reason)

Not for strict privacy.

It exists to **prevent accidental overriding in subclasses**.

---

### Example where it matters

```python
class Parent:
    def __init__(self):
        self.__data = "parent"

class Child(Parent):
    def __init__(self):
        super().__init__()
        self.__data = "child"

c = Child()

print(c._Parent__data)  # parent
print(c._Child__data)   # child
```

If mangling didn’t exist:

* `Child` would overwrite `Parent`’s variable silently → hard-to-debug bug

# PASS BY OBJECT REFERANCE

### in py a function can take any 'referance variable of an object' as its argument nad can also return any object referance ... 

````python
class Customer:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender


def greet(customer):
     ### customer is any referance variable of any Customer class
    if(customer.gender=="Female"): 
        print("hello miss",customer.name,"welcome back")   
    else:
        print("hello Sir",customer.name,"welcome back")              

cust = Customer("Kusumita","Female")
greet(cust)    
````

# COLLECTION OF OBJECTS 

````python
class Car:
    def __init__(self,colour,type):
        self.colour = colour
        self.type = type

suv1 = Car("blue","suv")
suv2 = Car("white","suv")
suv3 = Car("black","suv")
suv4 = Car("maroon","suv")

L = [suv1,suv2,suv3,suv4] ### collection of objects of Car class

for i in L:
    print(i.colour)
````
