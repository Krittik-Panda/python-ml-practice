### Instance variable is a variable which values are different for different objects of the same class .... in ATM_Class.ipynb file self.balance is an instance variable

#### use __ before any methode or instance variable to make it private so that one can not use those by using object ... if not make it private one can use it or even change it from object itself ...so for ATM class pin and balance should be private

```python
class ATM:
    def __init__(self):
        self.__pin = ""
        self.__balance = 0
        self.menu()
      
    def menu(self):        
        choice = int(input("""
            Enter the option no : 
            1) Set pin
            2) Check balance
            3) Money deposite
            4) Money withdraw
            5) Exit
            """) ) 
        if(choice==1):
            self.set_pin()
           # self.menu()
        elif(choice==2):
            self.check_balance()
            #self.menu()
        elif(choice==3):
            self.deposite()
            #self.menu()
        elif(choice==4):
            self.withdraw()
            #self.menu()     
        else:
            print("Thank You !")       
                

    def check_pin(self):
        temp = input("Enter the pin")
        if temp!=self.__pin:
            return 0
        else:
            return 1    


    def set_pin(self):
        self.__pin = input("Enter your pin ")
        print("Pin is successfully set ")


    def check_balance(self):
        if(self.check_pin()==0):
            print("Wrong pin ") 
        else:
            print(f"the balance of your current account is {self.__balance}")  


    def deposite(self):
        try:
            if(self.check_pin()==0):
                print("Wrong pin ")
            else:
                temp = int(input("Enter the deposite amount :"))
                self.__balance = self.__balance + temp
                print(f"your balance is {self.__balance}")
        except:
            print("Enter the digit properly !!!")            


    def withdraw(self):
        try:
            if(self.check_pin()==0):
                print("Wrong pin ")
            else:
                temp = int(input("Enter the withdraw amount : "))
                if(temp>self.__balance):
                    print("Insaficient Balance .")
                else:
                    self.__balance = self.__balance - temp
                    print("Withdrawn is successful ")
                    print(f"The current balance is {self.__balance}")    
        except:
            print("Enter digit properly !!!")  
                
```
