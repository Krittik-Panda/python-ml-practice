```python
import pandas as pd 
```

```python
indian_weather = pd.DataFrame({
    "city" : ["mumbai","delhi","banglore"],
    "temp" : [23,37,30],
    "humidirty" : [89,70,65]
}
)
us_weather = pd.DataFrame({
    "city" : ["new york","chicago","orlango"],
    "temp" : [13,34,23],
    "humidirty" : [81,72,45]
}
)

```

```python
df = pd.merge(indian_weather,us_weather,on = "temp")
df
### merge on the besis of temp , where temp values are same for these 2 data frame it merge these 2 data frame 
```

```python
indian_weather = pd.DataFrame({
    "city" : ["mumbai","delhi","banglore","balimore","oslo"],
    "humidirty" : [89,70,65,87,45]
}
)
us_weather = pd.DataFrame({
    "city" : ["delhi","banglore","mumbai","haldia"],
    "temp" : [13,34,23,787],
}
)
```

### inner join(default)

```python
df = pd.merge(indian_weather,us_weather,on = "city")
df
### merge perform inner join 
```

### outer join

```python
df = pd.merge(indian_weather,us_weather,on = "city",how = "outer")
df
```

### know in which data frame the data belongs to

```python
df = pd.merge(indian_weather,us_weather,on = "city",how = "outer",indicator= True)
df
```

```python

indian_weather = pd.DataFrame({
    "city" : ["mumbai","delhi","banglore"],
    "temp" : [23,37,30],
    "humidirty" : [89,70,65]
}
)
us_weather = pd.DataFrame({
    "city" : ["mumbai","delhi","banglore"],
    "temp" : [13,34,23],
    "humidirty" : [81,72,45]
}
)
```

```python
df = pd.merge(indian_weather,us_weather,on = "city" , how = "outer" , indicator=True , suffixes=("_left_table","_right_table"))
df
```
