### if the any column have any type of mixed data like 32C and 32F or wind speed 6mph or 7kmph etc 

```python
df = pd.read_csv("weather_data_2.csv")
df
```

```python

new_df = df.replace({
    'temperature': r'[A-Za-z]+',
    'windspeed': r'[A-Za-z]+'
}, np.nan, regex=True)
new_df
```

Here is a short 5–6 line explanation of these regex patterns:

* `'temperature': r'[A-Za-z]+'` and `'windspeed': r'[A-Za-z]+'` are **regular-expression rules** used in `df.replace()` to identify alphabet characters in the given columns.
* The pattern `[A–Z a–z]` matches **any uppercase or lowercase letter**.
* The `+` at the end means **one or more letters in sequence** (e.g., `C`, `mph`, `abc`, etc.).
* This allows pandas to detect **words or alphabetic units** inside numeric fields like `"30C"` or `"12mph"`.
* When used with `regex=True`, pandas can **search inside the text**, not just for exact matches.
* These matched alphabetic parts can then be **replaced with `np.nan` or blank** during data cleaning.
