# HANDLE MISSING VALUES

````python
import pandas as pd
import numpy as np
````

````python
df = pd.read_csv("weather_data.csv")
print(df)
````

````python
print(df.head(10))
````

````python
df.set_index('day',inplace=True)
print(df.head(20))
````

### fill all nan value with zero

````python

new_df = df.fillna(0.0)
print(new_df.head(20))
````

### fill all nan value with zero in a particular column 

````python
new_df = df.fillna({'temperature':0,'windspeed':0,'event':'no event'})
print(new_df.head(20))
````

### nan value is filled with the privious non nan value

````python
new_df = df.fillna(df.ffill())
print(new_df.head(10))
````

### nan value is filled with the next non nan value

````python
new_df = df.fillna(df.bfill())
print(new_df.head(10))
````

### copy horizontally left to right use ffill and axis = 1

````python
new_df = df.fillna(df.ffill(),axis=1)
print(new_df.head(10))
````

### copy horizontally right to left use bfill and axis = 1

````python
new_df = df.fillna(df.bfill(),axis=1)
print(new_df.tail(10))
````

### limit can limit how many times the value ffill(fordward fill) or bfill(backword fill) 

````python
print(df.head(10))

new_df = df.fillna(df.ffill(),limit = 1)
print(new_df.head(10))
````

````python
print(df.head(10))

new_df = df.fillna(method="ffill",limit = 1)
print(new_df.head(10))
````

Sure ‚Äî here is a **short difference + a big clear example** that shows the difference **visually**.

---

# üî• SHORT DIFFERENCE

| Code                                                          | Meaning                                                                  | Result                    |
| ------------------------------------------------------------- | ------------------------------------------------------------------------ | ------------------------- |
| `df.ffill(limit=1)`  or  `df.fillna(method="ffill", limit=1)` | Fill missing values **step-by-step**, only **1 NaN per consecutive gap** | Some NaN may remain       |
| `df.fillna(df.ffill(), limit=1)`                              | Replace NaN using values from a **fully forward-filled DataFrame**       | All NaN values get filled |

---

# üî• BIG EXAMPLE

### üìå Original DataFrame

```python
   A     B     C
0 NaN   10   NaN
1 5     NaN   x
2 6     NaN  NaN
3 7     20    y
4 NaN   30    z
5 9     NaN   t
```

---

### ‚úîÔ∏è Example 1 ‚Äî gradual forward fill

```python
df1 = df.ffill(limit=1)
print(df1)
```

üëâ Output:

```
     A     B    C
0   NaN   10   NaN
1   5.0   10     x
2   6.0  NaN     x
3   7.0   20     y
4   7.0   30     z
5   9.0   30     t
```

üìå Summary

* Only **one NaN per consecutive block** is filled
* Some NaNs remain (row 0 in A & C, row 2 in B)

---

### ‚úîÔ∏è Example 2 ‚Äî replacement using full forward-fill

```python
df2 = df.fillna(df.ffill(), limit=1)
print(df2)
```

üëâ Output:

```
     A     B    C
0   5.0   10     x
1   5.0   10     x
2   6.0   10     x
3   7.0   20     y
4   7.0   30     z
5   9.0   30     t
```

                   | Use `df.ffill(limit=1)` ‚Äî it‚Äôs the proper and future-safe syntax.         |


### Interpolation means estimating missing values using the trend of the existing data.

````python
df = df.astype({'temperature': float, 'windspeed': float})
new_df = df.interpolate()
print(new_df.head(20))


# d.astype() convert the temperature and windspeed to a floate value coz priviously it have the mixed data type so have the numeric value as a string 
````

````python
df.index = pd.to_datetime(df.index)


## index is explicitly converted to datetime dtype. previously it looks like 2025-01-01 but store as a string
new_df = df.interpolate(method="time")
print(new_df.head(20))
````

### drop the rows which have atleast one na value 

````python
new_df_1 = df.dropna()
print(new_df_1.head(20))
````

### want to drop the rows which have all na values

````python
new_df_1 = df.dropna(how="all")
print(new_df_1.head(20))
````

### if i have any row having atleast 1 non na value keep that row 

````python
new_df_1 = df.dropna(thresh=1)
print(new_df_1.head(20))
````

# fill missing index

````python
dt = pd.date_range("01-01-2025","06-01-2025")
indx = pd.DatetimeIndex(dt)
df = df.reindex(indx)
print(df)


#in this case no index is missing so nothing change but if the index are missing btw ("01-01-2025","06-01-2025") they all are incerted with NaN values
````
