```python
humidity_data = pd.DataFrame({
    "city" : ["mumbai","delhi","banglore"],
    "humidirty" : [89,70,65]
}
)
temp_data = pd.DataFrame({
    "city" : ["delhi","mumbai","banglore"],
    "temp" : [13,34,23],
}
)
```

```python
df = pd.concat([humidity_data,temp_data])
df
```

want to append column wise 


```python
df = pd.concat([humidity_data,temp_data],axis = 1)
df
```

order is misplaces not wrong but , missleading , include index parameter 

```python
humidity_data = pd.DataFrame({
    "city" : ["mumbai","delhi","banglore"],
    "humidirty" : [89,70,65]
},index = [0,1,2]
)
temp_data = pd.DataFrame({
    "city" : ["delhi","mumbai","banglore"],
    "temp" : [13,34,23],
},index = [1,0,2]
)
```

```python
df = pd.concat([humidity_data,temp_data],axis = 1)
df 
```

```python
df = pd.concat([humidity_data,temp_data],axis = 1)
df 
```

```python
df = pd.concat([humidity_data,temp_data],axis = 1,keys =["humidity","temperature"])
df 
```

### concate dataframe with series 

```python
s = pd.Series(["rain","humid","snowfall"],name = "condition")
df = pd.concat([temp_data,s],axis = 1)
df
```
