```python
df = pd.read_csv("data/weather_city_name.csv")
df
```

### **pivot table allows to resape or transform data**

```python
df1 = df.pivot_table(index = "city",columns = "date")
### its automatically return the mean where the values are same 
df1
```

```python
df1 = df.pivot_table(index = "city",columns = "date", aggfunc = "sum")
### return the sum of the values of same day
df1
```

```python
df3 = pd.read_csv("data/weather_same_city.csv")
df3 
```

### want to print the avd of the feb , mar and dec month using grouper function

```python
df3
df3['date'] = pd.to_datetime(df3['date'], dayfirst=True)
df3
```

```python
df2 = df3.pivot_table(index = pd.Grouper(key="date",freq="M"),columns = "city")
# M means the month , here in the data frame there are 3 months ... feb ,mar,dec ,,,grouper func will return the avg humidity and temp of feb,mar,dec 
df2
```
