```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "Day" : ["sunday","monday","tuesday","wednesday","thrusday","friday","saturday"],
        "chicago" :[23,45,23,41,23,45,44],
        "barlin" : [30,45,44,43,40,41,42],
        "new york" : [23,22,23,24,23,21,22]
    }
)
df
```

numbers are temperatures ...we wnat to resahpe the dataframe into day , city tempereture

```python
df1 = pd.melt(df,id_vars=["Day"])
df1
```

variable --> city , value --> temperature

```python
df2 = pd.melt(df,id_vars=["Day"],var_name="CITY",value_name="TEMPERETURE",ignore_index=True)
df2
```
