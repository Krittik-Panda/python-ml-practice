| Feature   | Series                      | DataFrame                           |
| --------- | --------------------------- | ----------------------------------- |
| Dimension | 1-D                         | 2-D                                 |
| Structure | Single column               | Multiple columns (even if only one) |
| Display   | index + values              | rows + columns                      |
| Type      | `pandas.core.series.Series` | `pandas.core.frame.DataFrame`       |


```python
df_1 = pd.Series([1,2,3,4])
print(df_1)
print(type(df_1))
df_2 = pd.DataFrame([1,2,3,4])
print(df_2)
print(type(df_2))
print(np.sum(df_1))
print(df_1 + df_2 )
```

### To add (sum) two Pandas Series element-by-element using NumPy, you can do it very easily.

```python

s1 = pd.Series([1, 2, 3, 4])
s2 = pd.Series([10, 20, 30, 40])

result = np.add(s1, s2)     # element-by-element sum
print(result)
```
